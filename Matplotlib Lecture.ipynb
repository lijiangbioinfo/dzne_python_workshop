## Matplotlib Overview

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([2, 3, 4, 3, 4]);
plt.ylabel("Y")
plt.xlabel("X")
plt.title("My Plot")
plt.xticks([0, 1, 2, 3, 4]);
```

```python
plt.plot([2, 3, 4], [10, 15, 5])
```

```python
plt.plot([2, 3, 4], [10, 15, 5], c=[1., 0., 0.])
```

```python
plt.scatter(x=np.random.randn(50), y=np.random.random(50), c='red', s=500, alpha=.5)
```

```python
y = np.random.random(50)

plt.scatter(
    x=np.random.randn(50), 
    y=y,
    c='red', 
    s=y*500, 
    alpha=.5
)
plt.title("Title");
```
